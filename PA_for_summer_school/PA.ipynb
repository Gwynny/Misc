### Задача:
Пьяница стоит на краю утёса на расстоянии __двух шагов__ и делает случайные
шаги вперёд-назад (шаг назад с вероятностью __2/3__, шаг вперёд с вероятностью
__1/3__). Какова вероятность того, что после __5 шагов__ пьяница не свалился?
Напишите функцию, принимающую N и возвращающую вероятность того,
что после этих N шагов пьяница жив. Постройте график этой функции.

```python
def func(N):
    q = []
    for i in range(10000):
        a = 2
        a += sum(np.random.choice([1, -1], size = int(N), p = [2.0/3, 1.0/3]))
        if (a == 0) | (a < 0):
            q.append(0)
        else:
            q.append(1)
    return float(sum(q))/len(q)
```

Вероятность того,что он будет жив после пяти шагов

```python
func(5)
```

```python
x = np.arange(2, 100, 2)
y = []
for i in range(len(x)):
    y.append(func(x[i]))
```

```python
fig = plt.figure(figsize=(10,8))
plt.plot(x, y)
fig.suptitle('Probability to stay alive after N steps', fontsize=20)
plt.xlabel('Number of steps', fontsize=16)
plt.ylabel('Probability', fontsize=16)
```
